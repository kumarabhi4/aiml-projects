Q 2. An electrical manufacturing company conducts quality checks at specified periods on the products it manufactures. Historically, the failure rate for the manufactured item is 5%. Suppose a random sample of 10 manufactured items is selected. Answer the following questions.

```python
p = 0.05
n = 10

#2.A. Probability that none of the items are defective? 
k = 0
P1 = stats.binom.pmf(k, n, p)
print('2.A. Probability that none of the items are defective is %1.4f' %P1)

#2.B. Probability that exactly one of the items is defective? 
k = 1
P2 = stats.binom.pmf(k, n, p)
print('2.B. Probability that exactly one of the items is defective is %1.4f' %P2)

#2.C. Probability that two or fewer of the items are defective? 
k = np.arange(0,3)
cb = stats.binom.cdf(k, n, p)
P3 = cb[2]
print('2.C. Probability that two or fewer of the items are defective is %1.4f' %P3)


#2.D. Probability that three or more of the items are defective ?
P4 = 1-P3
print('2.D. Probability that three or more of the items are defective is %1.4f' %P4)
```

Q 3. A car salesman sells on an average 3 cars per week.

```python
rate = 3
n = np.arange(0,10)
poisson = stats.poisson.pmf(n,rate)
#3.A. What is Probability that in a given week he will sell some cars? 
P1 = 1-poisson[0]
print('3.A. Probability that in a given week he will sell some cars is %1.4f' %P1)
```

```python
#3.B. What is Probability that in a given week he will sell 2 or more but less than 5 cars? 
P2 = poisson[1] #probability to sell 2 cars
P3 = poisson[2] #probability to sell 3 cars
P4 = poisson[3] #probability to sell 4 cars

P = P2+P3+P4
print('3.B. Probability that in a given week he will sell 2 or more but less than 5 cars is %1.4f' %P)
```

```python
#3.C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week.
rate = 3
x = np.arange(0, 11)

poisson_cdf = stats.poisson.cdf(x, rate)

plt.figure(figsize=(10, 6))
plt.plot(x, poisson_cdf)
plt.title('poisson distribution')
plt.xlabel('number of cars sold per week')
plt.ylabel('cumulative probability of cars sold per-week')
plt.grid(True)
plt.show()
```

**4.B. Write the null hypothesis and alternative hypothesis.**

**Null Hypothesis H0:** There is no difference between Funds raised by companies that are still operating vs company that are closed now.

**Alternative Hypothesis H1:** There is difference between Funds raised by companies that are still operating vs company that are closed now.    

**4.C. Test for significance and conclusion**



```python
import numpy                     as     np
import pandas                    as     pd
from   scipy.stats               import ttest_1samp, ttest_ind
import matplotlib.pyplot         as     plt
import matplotlib
import seaborn as sns
import scipy.stats as stats
import statsmodels.stats.api as sm

#Alpha - 0.05

#STATISTICAL TEST

newdf1 = newdf[['OperatingState','Funds_in_million']]
newdf2 = newdf1.loc[newdf1['OperatingState']=='Closed']
newdf3 = newdf1.loc[newdf1['OperatingState']=='Operating']

#Calculate p value

t_statistic, p_value  = ttest_ind(newdf2['Funds_in_million'],newdf3['Funds_in_million'])

#print('tstat',t_statistic)    
#print('P Value',p_value)

# p_value < 0.05 => alternative hypothesis:
# they don't have the same mean at the 5% significance level
print ("two-sample t-test p-value=", p_value)

alpha_level = 0.05

if p_value < alpha_level:
    print('We have enough evidence to reject the null hypothesis in favour of alternative hypothesis')
    print('We conclude that there is significant difference between Funds raised by companies that are still operating vs company that are closed now.')
else:
    print('We do not have enough evidence to reject the null hypothesis in favour of alternative hypothesis')
    print('We conclude that there is no significant difference between Funds raised by companies that are still operating vs company that are closed now.')
```
